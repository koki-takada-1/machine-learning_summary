```python
import pandas as pd
from sklearn.datasets import load_wine

data = load_wine()

#Pandasを用いて特徴量とカラム名を取り出す
data_x = pd.DataFrame(data = data.data, columns = data.feature_names)

#データがもつ特徴量を上から5行表示
display(data_x.head())

#Pandasを用いてラベルを取り出す
data_y = pd.DataFrame(data = data.target)
display(data_y)


#カラム名が「0」となっていてわかりずらいので「class」に変更してみる
data_y = data_y.rename(columns = {0:'class'})

#データに割り振られたラベルを上から5行表示
display(data_y.head())
```

```python
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split

#ワインのデータをインポート
wine = load_wine()

#特徴量とラベルデータを抽出
data = wine.data
target = wine.target

#データを分割
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size = 0.2, random_state = 0)

#決定木
clf = tree.DecisionTreeClassifier()

#学習
clf = clf.fit(x_train, y_train)

#正解率
print(clf.score(x_test, y_test))
```

```python
import pydot
```

```python
tree.export_graphviz(
    clf,
    feature_names = wine.feature_names,
    class_names = wine.target_names,
    filled = True,
    rounded = True,
    out_file = "tree.dot"
)

#tree.dotを視覚化して、画像で出力
(graph, )=pydot.graph_from_dot_file("tree.dot")
graph.write_png("tree.png")
```
